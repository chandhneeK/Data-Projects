```python
from bs4 import BeautifulSoup 
import re 
import urllib.request as urllib2
import urllib 
import urllib.error
import pandas as pd 
import numpy as np
import os
import requests
os.environ['http_proxy']=''
# read urls of websites from text file 
browser_history = pd.read_csv('/content/domains - bq-results-20220427-195415-1651089299245 (1).csv')
browser_history = browser_history.dropna()

urls = browser_history['value'].to_list()
browser_history.head()


```

```python
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36'
#user_agent = 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.102 Safari/537.36'
#user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
c_headers={'User-Agent':user_agent} 
conn_timeout = 6
read_timeout = 60
timeouts = (conn_timeout, read_timeout)
parsed_text = []
for url in urls: 

    try:
        response = 'no title'
        response = requests.get(url,timeout = timeouts, headers = c_headers)
        response.raise_for_status()
        
        # Code here will only run if the request is successful
    except requests.exceptions.HTTPError as errh:
        print(errh)
    except requests.exceptions.ConnectionError as errc:
        print(errc)
    except requests.exceptions.Timeout as errt:
        print(errt)
    except requests.exceptions.RequestException as err:
        print(err)
    if response != 'no title':    
        soup = BeautifulSoup(response.content, "html5lib")    
    if soup.title != None and response != '':
        parsed_text.append(soup.title.string)
        soup.title = None
    else: 
        parsed_text.append('no title')
    
        
    
        
```

```python
output = pd.DataFrame(parsed_text)
```

```python
output.to_csv('domain_titles.csv', index=False)
```
