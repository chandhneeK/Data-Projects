```python


books['Genre_bin'] = np.where(books['Genre']=='Fiction',1,0)
print (books.head())
```

```python
X = np.arange(6)
fig = plt.figure()
ax = fig.add_axes([10,10,3.5,1])


plt.scatter(books['Reviews'], books['Price'], c=books['Year'], s= books['User Rating']*10, alpha= 0.5 )
cbar = plt.colorbar(ticks =[0,1] )
cbar.set_ticklabels(["Non-Fiction", "Fiction"])
```

```python
Genre = books.loc[:,['User Rating','Reviews','Price','Year']].groupby(['Year']).mean()
print (Genre)
```

```python
Genre = books.loc[:,['User Rating','Price','Year','Reviews']].groupby(['Year']).mean()
Genre.plot(kind = "bar", figsize= (20,10), logy = True)
plt.show()
```

```python
fig0, ax0 = plt.subplots()
ax1 = ax0.twinx()

Genre = books.loc[:,['User Rating','Price','Year','Reviews']].groupby(['Year']).mean()
Genre.plot(kind = "bar", figsize= (20,10), ax=ax0)
books.loc[:,['User Rating','Price','Year']].groupby(['Year']).mean().plot(kind="bar", ax=ax1 , alpha =0.7)
plt.show()
```

```python
Genre = books.loc[:,['User Rating','Price','Year','Reviews']].groupby(['Year']).mean()
Genre.plot(kind = "line", figsize= (20,10), logy = True)
plt.show()
```

```python
#Top Books in Fiction

books.loc[books.loc[:,'Genre']=='Fiction',:].sort_values(by=['Year','User Rating', 'Reviews'], axis=0, ascending = False).head(20)
```

```python
#Top Books in Non-Fiction

books.loc[books.loc[:,'Genre']!='Fiction',:].sort_values(by=['Year','User Rating', 'Reviews'], axis=0, ascending = False).head(20)
```
