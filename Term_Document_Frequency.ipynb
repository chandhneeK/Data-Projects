```python
#importing necessary libraries
import nltk
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nltk.download('punkt') # tokenize
nltk.download('wordnet') # lemmatize
nltk.download('stopwords') #stopwords
nltk.download('averaged_perceptron_tagger') # POS
```

```python
#reading document

c1 = pd.read_csv('/content/IA2.csv', index_col=0, header= None)
c1.head()
```

```python
# tokenize

c1_tokenized = []
for review in c1.iloc[:,0]:
  c1_tokenized.append(nltk.word_tokenize(review))

#c1_tokenized = pd.DataFrame(c1_tokenized)
print(c1_tokenized)

```

```python
# lemmatize
lemmatizer = nltk.stem.WordNetLemmatizer()
c1_lemmatized =[]

x= len(c1_tokenized)
for review in c1_tokenized:
  t=[]
  for token in review:
    if token.isalpha():
      t.append(lemmatizer.lemmatize(token))
  c1_lemmatized.append(t)      
print(c1_lemmatized)
```

```python
# remove stop words
from nltk.corpus import stopwords

c1_stop =[]

for review in c1_lemmatized:
  t=[token for token in review if not token.lower() in stopwords.words('english')]
  c1_stop.append(t)      
print(c1_stop)
```

```python
#converting the right format for further processing

master = [' '.join(review) for review in c1_stop]
print(master)
```

```python
#tfidf


tfidf_master = TfidfVectorizer(ngram_range=(1, 2),min_df=3) # remove low frequency words
tfidf_master.fit(master)
tfidf_final = tfidf_master.transform(master)


#getting vocabulary
data_items = tfidf_master.vocabulary_.items()
data_list = list(data_items)
df = pd.DataFrame(data_list, columns = ['review', 'place'] )
df.sort_values('place', inplace = True)
df.set_index('place', inplace = True)


output = pd.DataFrame(tfidf_final.toarray(), columns = df['review'])
output.to_csv(r'/content/output.csv', index=False)
print(output.shape)
output.head()
```

```python
#tfidf with POS tag

c2 = [' '.join(review) for review in c1_tokenized]

POS_c2 = []

for doc in c2:
    token_doc = nltk.word_tokenize(doc)
    POS_token_doc = nltk.pos_tag(token_doc)
    temp = [] # list of concatenated token with POS tag
    for i in POS_token_doc:
        temp.append(i[0] + "_" + i[1])
    POS_c2.append(" ".join(temp))

tfidf_master = TfidfVectorizer(min_df=4) # remove low frequency words
tfidf_master.fit(POS_c2)
c2_tfidf_final = tfidf_master.transform(POS_c2)


#getting vocabulary
data_items = tfidf_master.vocabulary_.items()
data_list = list(data_items)
df = pd.DataFrame(data_list, columns = ['review', 'place'] )
df.sort_values('place', inplace = True)
df.set_index('place', inplace = True)


output_pos = pd.DataFrame(c2_tfidf_final.toarray(),  columns = df['review'])
output_pos.to_csv(r'/content/output_pos.csv', index=False)
print(output_pos.shape)
output_pos.head()
```
