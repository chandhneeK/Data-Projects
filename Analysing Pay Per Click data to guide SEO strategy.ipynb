```python
#Question5:
#The company expects that the exact match and phrase match keywords would provide higher ROA 
#compared to broad match keywords. While updating their strategy they are thinking about bidding 
#on only exact and phrase match keywords. Would you support this decision? Why/Why Not? 

#Calculate ROA
df= bgame.copy()
def roa(x):
  if x['Total Advertising Cost']==0:
    return 0
  else:
    return x['Total Revenue']/x['Total Advertising Cost']  

df['Roa'] = df.apply(lambda x: roa(x),axis=1) 

#Aggregating on Keyword Type
keywords = pd.pivot_table(df, values = ['Avg Position', 'Total Advertising Cost', 'Total Revenue','Roa'], index = ['Keyword Type'], aggfunc = np.mean)
print(keywords)
#ROA by Keyword Type
sns.barplot(x='Keyword Type', y='Roa', data = df)
plt.xlabel('Keyword Type')
plt.ylabel('Average ROA')
plt.title('How does Average ROA Vary by Keyword type?')
plt.show()

```

```python
#6 The company feels that global audience (targeted through Google -Global, MSN Global and Yahoo - Global) 
# does not purchase the products from it and wishes to know if it is worthwhile to still invest in advertising 
# to global audience. What do you recommend?

df = pd.read_csv("https://raw.githubusercontent.com/dpham89/MSIS-521-Team-2/main/TA1.csv")
df_global = df.loc[(df['Publisher'] == 'Google-Global') | (df['Publisher'] == 'MSN-Global') | (df['Publisher'] == 'Yahoo-Global')]
table_merged = pd.pivot_table(df_global, values = ['Clicks', 'Impressions', 'Total Advertising Cost','Conversions','Revenue per Conversion', 'Total Revenue'], 
                       index = ['Publisher'], aggfunc = np.mean)
table_merged['ROA'] = table_merged['Total Revenue']/table_merged['Total Advertising Cost']
table_merged['CTR'] = table_merged['Clicks']/table_merged['Impressions']
table_merged['TCR'] = table_merged['Conversions']/table_merged['Clicks']       
print('Average Stats for Global Advertising')
display(table_merged.sort_values('ROA', ascending=0))

print('\n \n')

#US
df_us = df.loc[(df['Publisher'] == 'Google-US') | (df['Publisher'] == 'MSN-US') | (df['Publisher'] == 'Yahoo-US')]
table_merged2 = pd.pivot_table(df_us, values = ['Clicks', 'Impressions', 'Total Advertising Cost','Conversions','Revenue per Conversion', 'Total Revenue'], 
                       index = ['Publisher'], aggfunc = np.mean
                      )
table_merged2['ROA'] = table_merged2['Total Revenue']/table_merged2['Total Advertising Cost']
table_merged2['CTR'] = table_merged2['Clicks']/table_merged2['Impressions']
table_merged2['TCR'] = table_merged2['Conversions']/table_merged2['Clicks']
print('Average Stats for US Advertising')
display(table_merged2.sort_values('ROA', ascending=0))
```

```python
#QUESTION7
#Would the search side strategies that the company should employ be different for US audience and
#Global audience? Please explain what the strategies should be.

#We will view the investment for US audience and Global audience
table = pd.pivot_table(bgame, values = ['Clicks','Impressions', 'Total Advertising Cost','Conversions','Revenue per Conversion', 'Total Revenue'], 
                       index = ['Publisher'], aggfunc = np.sum)
```

```python
table_cpc = pd.pivot_table(bgame, values = ['Avg CPC'], index = ['Publisher'], aggfunc= np.mean)
table_n = pd.concat([table, table_cpc], axis = 1)
```

```python
table_n['ROA'] = table_n['Total Revenue']/table_n['Total Advertising Cost']
table_n['CTR'] = table_n['Clicks']/table_n['Impressions']
table_n['Prob. Booking'] = table_n['Conversions']/table_n['Impressions']
table_n['Sum Click Charges'] = table_n['Avg CPC']* table_n['Clicks']
table_n.sort_values('ROA', ascending=0)
```

```python
import matplotlib.pyplot as plt

x = table_n['Prob. Booking'] * 1000
y = table_n['Avg CPC'] 
size = table_n['Sum Click Charges']
color = np.random.rand(7) 

glb_avg_prob_booking = np.sum(table_n['Conversions']) / np.sum(table_n['Impressions'])
glb_avg_cost_click = np.sum(table_n['Sum Click Charges']) / np.sum(bgame['Clicks']) 

print(y)
```

```python
plt.figure(figsize=(10,6))
plt.axis([0, 15, 0, 1])
plt.xlabel('Prob. Booking (x 1000)')
plt.ylabel('Avg. CPC')
plt.title('Total Investment')

plt.scatter(x, y, s = size, c = color)

plt.plot([0, 15], [glb_avg_cost_click, glb_avg_cost_click]) # globaal avg CPC
plt.plot([glb_avg_prob_booking * 1000, glb_avg_prob_booking * 1000], [0, 1])

# add label to each bubble
for xi, yi, i in zip(x, y, x.index):
    plt.annotate(i, (xi, yi), textcoords = "offset pixels", xytext = (0, 10), ha = 'center')

plt.show()
```
