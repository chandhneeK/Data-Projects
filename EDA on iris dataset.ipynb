```python
from sklearn.datasets import load_iris
iris = load_iris()
attr = iris.data
print (iris.feature_names)
iris.data
```

```python
plt.scatter(attr[:,0], attr[:,1],c = 'g')

plt.xlabel("sepal length",size = 12)
plt.ylabel("sepal width", size = 12)
plt.title ("Correlation between sepal length and sepal width")
plt.show()
```

```python
target = iris.target
plt.scatter(attr[:,0], attr[:,1],c = target, s = attr [:,2]*10)

plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.title ("Correlation between sepal length and sepal width")
plt.show()
```

```python
select_index = target == 0
setosa = attr [select_index,:]
plt.hist (setosa [:,1])
plt.show()
```

```python
import pandas as pd 
daily=pd.read_csv('day.csv')
daily.head()
```

```python
plt.plot(daily.loc[:,"day"], daily.loc[:,"cnt"])
plt.xlabel ("Day")
plt.ylabel ("checkouts")
plt.xticks (range (min(daily.loc[:,"day"]), max(daily.loc[:,"day"]),25))
plt.figure(figsize= (0,1))
plt.show()

```

```python
plt.scatter(daily.loc[:,"temp"],daily.loc[:,"casual"], c = '#539cab' , alpha = 0.7 )
plt.ylabel ("Unregistered Checkout count")
plt.xlabel ("Temperature")
plt.show()

```

```python
plt.scatter(daily.loc[:,"temp"],daily.loc[:,"casual"],c = daily.loc[:,"workingday"] , alpha = 0.7 )
plt.ylabel ("Unregistered Checkout count")
plt.xlabel ("Temperature")
cbar = plt.colorbar(ticks =[0,1] )
cbar.set_ticklabels(["Not a workday", "Workday"])
plt.show()
```

```python
plt.hist(daily.loc[:,"registered"],10)
plt.xlabel("Number of Observations")
plt.ylabel("Resigstered Checkouts")
plt.title("Distribution of Registered Checkouts")
plt.show()
```

```python
plt.hist(daily.loc[:,"registered"],10)
plt.hist(daily.loc[:,"casual"],5, alpha = 0.8)
plt.xlabel("Number of Observations")
plt.ylabel("Resigstered Checkouts & Unregistered Checkouts")
plt.title("Distribution of Registered and Unregistered Checkouts")
plt.show()
```

```python
import numpy as np
plt.scatter( daily.loc[:,"temp"],daily.loc[:,"season"])
plt.yticks([1,2,3,4])
print(np.mean(daily.loc[:,"temp"]))
print (np.median(daily.loc[:,"temp"]))
```

```python
winter = daily.loc[daily.loc[:,"season"]==1,'temp']
spring = daily.loc[daily.loc[:,"season"]==2,'temp']
summer = daily.loc[daily.loc[:,"season"]==3,'temp']
fall = daily.loc[daily.loc[:,"season"]==4,'temp']

plt.hist(winter, 10, alpha =0.7)
plt.hist(spring, 10,alpha =0.7)
plt.hist(summer, 10,alpha =0.7)
plt.hist(fall, 10, alpha =0.7)
plt.plot(np.mean(daily.loc[:,"temp"],[0:50])
plt.show()
```
